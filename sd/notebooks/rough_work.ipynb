```python
import matplotlib.pyplot as plt
import torch

plt.figure(figsize=(10,10))

beta_start = 0.00085
beta_end = 0.0120
num_training_steps = 1000

betas = torch.linspace(beta_start ** 0.5, beta_end ** 0.5, num_training_steps, dtype=torch.float32) ** 2
alphas = 1 - betas
alpha_cumprod = torch.cumprod(alphas, dim=0)

plt.subplot(3, 1, 1)
plt.plot(betas)
plt.title("betas")


plt.subplot(3, 1, 2)
plt.plot(alphas)
plt.title("alphas")


plt.subplot(3, 1, 3)
plt.plot(alpha_cumprod)
plt.title("alpha_cumprod")
```
